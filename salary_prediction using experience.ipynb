```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color="yellow")
plt.title("Training Data")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color="yellow")
plt.title("Testing Data")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
data = pd.read_csv('Salary_Data.csv')
X = data[['YearsExperience']]
y = data['Salary']
lr_poly2= LinearRegression()
lr_poly2.fit(X, y)
joblib.dump(lr_poly2, 'lr_poly2.joblib')
```

```python
lr_single_feature = joblib.load('lr_poly2.joblib')
num_features = lr_single_feature.coef_.shape[0] if hasattr(lr_single_feature, 'coef_') else None
print(f"Number of features in the retrained model: {num_features}")
```
